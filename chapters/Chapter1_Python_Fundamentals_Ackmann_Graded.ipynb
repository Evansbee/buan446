### Important Type Behaviors & Dynamic Typing

Python is a **dynamically typed** language. This means:

1. **Variables don't have a fixed type**: You can assign an integer to a variable `x` now, and a string to `x` later (though it's often confusing to do so).
2. **Types are attached to values, not variables**: `10` is an integer, `"10"` is a string. The variable just points to the value.
3. **Operations depend on the value's type**: The meaning of `+` changes depending on whether you are adding numbers or strings.

Let's see how different types behave differently with the same operators:

````python
# Addition with numbers
print(105 + 15)        # Mathematical addition: total credits

# Addition with strings (concatenation)
print("CU" + "100001")    # String joining: building an ID
````

````python
# Multiplication with numbers
print(3 * 15)   # 3 semesters x 15 credits each

# Multiplication with strings (repetition)
print("=" * 40)  # Create a divider line by repeating the provided string 40 times
````

## 1.4 Arithmetic Operators

Python supports standard mathematical operations:

| Operator | Operation | Example | Result |
|----------|-----------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `5 / 3` | `1.666...` |
| `//` | Floor Division | `5 // 3` | `1` |
| `%` | Modulo (remainder) | `5 % 3` | `2` |
| `**` | Exponentiation | `5 ** 3` | `125` |

### Understanding Floor Division (`//`) and Modulo (`%`)

These two operators are often new to beginners but are extremely useful in data analysis.

**Floor Division (`//`)**:
- Also called "integer division".
- It divides two numbers and rounds the result *down* to the nearest whole number.
- **Use case**: Checking how many full groups fit into a total.
  - *Example*: If you have 105 students and want to form groups of 4, `105 // 4` tells you how many full groups you can make.

**Modulo (`%`)**:
- Returns the **remainder** of a division.
- **Use case**: Finding what's left over, or checking for patterns.
  - *Example*: `105 % 4` tells you how many students are left without a group.
  - *Example*: `number % 2` is a standard way to check if a number is even (result is 0) or odd (result is 1).

````python
# Division always returns a float
print(100 / 4)       # 25.0, not 25

# Floor division truncates to integer
print(105 // 15)     # How many full semesters of 15 credits?

# Modulo gives the remainder as an integer
print(105 % 15)      # Credits left over after full semesters
````

````python
# Practical example: calculating credit completion rate
credits_attempted = 105
credits_earned = 99

completion_rate = credits_earned / credits_attempted
print(completion_rate)

# As a percentage
completion_percentage = (credits_earned / credits_attempted) * 100
print(completion_percentage)
````

````python
# Calculating average GPA across students
total_gpa_points = 3.41 + 3.55 + 3.60 + 3.44 + 3.90
student_count = 5

average_gpa = total_gpa_points / student_count
print(average_gpa)
````

### Operator Precedence

Python follows standard mathematical order of operations (PEMDAS):

1. Parentheses `()`
2. Exponentiation `**`
3. Multiplication, Division, Floor Division, Modulo `* / // %`
4. Addition, Subtraction `+ -`

When in doubt, use parentheses to make your intent clear.

````python
# Without parentheses
result1 = 2 + 3 * 4 # could be written 2 + (3 * 4) for clarity
print(result1)  # 14, not 20

# With parentheses if you want to add the numbers before multiplying by 4
result2 = (2 + 3) * 4
print(result2)  # 20
````

````python
# GPA calculation example
# Quality points = credits * grade points
# GPA = total quality points / total credits

credits_course1 = 3
grade_course1 = 4.0  # A

credits_course2 = 4
grade_course2 = 3.0  # B

# This needs parentheses to work correctly!
gpa = (credits_course1 * grade_course1 + credits_course2 * grade_course2) / (credits_course1 + credits_course2)
print(gpa)
````

## 1.7 F-Strings (Formatted String Literals)

F-strings provide a clean way to embed expressions inside strings. Start the string with `f` and put expressions inside `{}`.

**Reference:** For a comprehensive guide, check out [f-string help](https://fstring.help/).

````python
student_id = "CU100001"
gpa = 3.41
college = "College of Business"

# Without f-string (awkward)
# Remember when we use `+` with strings it concatenates (puts them together)
print("Student " + student_id + " has a GPA of " + str(gpa))

# With f-string (clean)
# Just one set of quotes at the beginning and end of the string and variable are in curly braces
print(f"Student {student_id} has a GPA of {gpa}")
````

### Format Specifiers

F-strings support format specifiers after a colon inside the braces. These are essential for professional data reporting and another advantage of f-strings:

````python
credits_attempted = 105
credits_earned = 99
completion_rate = credits_earned / credits_attempted

# .2f means "2 decimal places, float format"
print(f"Credits earned: {credits_earned}")
print(f"Completion rate: {completion_rate:.2f}")  # 0.94
print(f"Completion rate: {completion_rate:.1%}")  # 94.3% (percentage format!)
````

````python
# Formatting student data for reports
gpa = 3.4142857
total_students = 600

print(f"Average GPA: {gpa:.2f}")           # 3.41
print(f"Total enrollment: {total_students:,}")  # 600 (would show 1,234 for larger numbers - try it out)
print(f"GPA (3 decimals): {gpa:.3f}")      # 3.414
````

````python
# Creating a formatted student report
student_id = "CU100001"
major = "Accounting"
gpa = 3.41
credits = 105

print("=" * 40)
print("STUDENT RECORD")
print("=" * 40)
print(f"ID:      {student_id}")
print(f"Major:   {major}")
print(f"GPA:     {gpa:.2f}")
print(f"Credits: {credits}")
print("=" * 40)
````

## Chapter Project: Student Academic Metrics Calculator

Build an interactive tool that calculates key academic metrics for a student. This mirrors what an academic advisor might use.

### Requirements

**The program should:**
1. Collect student information via input:
   - Student ID
   - Current GPA
   - Credits attempted
   - Credits earned

2. Calculate and display:
   - Credit completion rate (as percentage)
   - Quality points (GPA × credits earned)
   - Approximate semesters completed (credits ÷ 15)
   - Credits needed to graduate (assuming 120 required)

3. Generate a formatted report that looks professional

**Bonus challenges:**
- Add Dean's List status (GPA ≥ 3.5 with at least 12 credits)
- Calculate what GPA is needed in remaining credits to reach 3.0 or 3.5 overall
- Handle the case where the student has already met graduation requirements

### Working with AI

Effective prompts to consider:
- "How do I format numbers as percentages in Python f-strings?"
- "What's the formula to calculate required GPA in remaining credits to achieve a target GPA?"
- "How can I make my printed output align in columns?"

Avoid prompts like:
- "Write a student metrics calculator for me"
- "Complete this assignment"

The goal is to use AI as a knowledgeable colleague, not as someone who does your work.

````python
# STUDENT ACADEMIC METRICS CALCULATOR
# 
# Document your AI collaboration below:
# - Prompts used: in a jupyter notebook, using python, i want to input multiple students worth of data.  I have an input("Do you want to do more y/n") at the end, is this the best way to handle it in the notebook framework, is there something cute that jupyter would interpret as a button for yes/no?
# - Key insights from AI: the "ipywidgets" library exists and that's super cool, but i'm not going to do that right now...
# - Modifications you made: none, though i'm going to explore those ipywidgets features... that's freaking cool
#
# Your code below:

#some magic numbers so we don't have to dig if requirements change
TOTAL_CREDITS = 120
DEANS_LIST_GPA = 3.5

#the future gpa is really a quality points question
REQUIRED_QUALITY_POINTS = TOTAL_CREDITS * DEANS_LIST_GPA


print("=" * 80)
print("CRESTVIEW UNIVERSITY - STUDENT METRICS CALCULATOR")
print("=" * 80)

#array to keep track of all the students we're going to enter later
students = []


#easy dict to allow for column renaming if something changes or we want to make it shorter
col_key_to_name = {
    'id': 'Student ID',
    'gpa': 'GPA',
    'cr_attempted': 'Credits Attempted',
    'cr_earned': 'Credits Earned',
    'completion_rate': 'Comp. Rate',
    'quality_points': 'Quality Points',
    'semesters_complete': 'Semesters Complete',
    'credits_remain':'Credits Remaining',
    'deans_list': 'Deans List',
    'req_gpa_remain':'Required GPA'
}

#padding separate from column length is a design decision, but i like it.
PADDING = 2

#this keeps track of the column widths for all the fields.  A smarter version of this would loop through the data to see if any 
#data in the fields would require a wider entry, but we're not there right now.  The headers are longer than the data, won't always be true
# (see gpa below)
col_key_to_width = {
    'id': len(col_key_to_name['id']),
    'gpa': len(col_key_to_name['gpa']) + 1, #gpas are 4 characters and i don't want to dynamically size columns on the first assignment
    'cr_attempted': len(col_key_to_name['cr_attempted']),
    'cr_earned': len(col_key_to_name['cr_earned']),
    'completion_rate': len(col_key_to_name['completion_rate']),
    'quality_points': len(col_key_to_name['quality_points']),
    'semesters_complete': len(col_key_to_name['semesters_complete']),
    'credits_remain':len(col_key_to_name['credits_remain']),
    'deans_list': len(col_key_to_name['deans_list']),
    'req_gpa_remain' :len(col_key_to_name['req_gpa_remain']),
}

#loop until break
while True:
    #new dict for every lovely student
    student = {}
    student['id'] = input("Enter Student ID: ").upper() #so that su -> SU for consistency
    student['gpa'] = float(input("Enter GPA: "))
    student['cr_attempted'] = int(input("Enter Credits Attempted: "))
    student['cr_earned'] = int(input("Enter Credits Earned: "))
    student['completion_rate'] = student['cr_earned'] / student['cr_attempted']
    student['quality_points'] = student['gpa'] * student['cr_earned'] 
    student['semesters_complete'] = student['cr_earned'] // 15
    student['credits_remain'] = max(TOTAL_CREDITS - student['cr_earned'],0)
    student['deans_list'] = student['gpa'] >= 3.5 and student['cr_earned'] >= 12


    #this is a bad thing in general but given this is a report, i'm actually going to stringify these
    #now, that way I can just loop print everythign later and it'll work nice, also allows me to dynamically 
    #calculate the required field width if i wanted to do that...
    
    
    
    #Do this first as a string to cut things out
    if student['credits_remain'] != 0:
        student['req_gpa_remain'] = (REQUIRED_QUALITY_POINTS - student['quality_points']) / student['credits_remain']
        if student['req_gpa_remain'] > 4.0 or student['req_gpa_remain'] < 0:
            student['req_gpa_remain'] = 'N/A'
        else:
            student['req_gpa_remain'] = f"{student['req_gpa_remain']:.2f}"
    else: 
        student['req_gpa_remain'] = 'N/A'

    #also nice because we only have to format here.
    student['gpa'] = f"{student['gpa']:.2f}" 
    student['cr_attempted'] = f"{student['cr_attempted']}" 
    student['cr_earned'] = f"{student['cr_earned']}" 
    student['completion_rate'] = f"{student['completion_rate']:.2%}" 
    student['quality_points'] = f"{student['quality_points']:.2f}" 
    student['semesters_complete'] = f"{student['semesters_complete']}" 
    student['credits_remain'] = f"{student['credits_remain']}" 
    if student['deans_list']:
        student['deans_list'] = "✓"
    else:
        student['deans_list'] = "✗"
    
    students.append(student)
    if input("More entries? [y/N]: ").lower() != 'y':
        break

#let it breath!
print()
#print headers
printed_width = 0
for key,header in col_key_to_name.items():
    print(f"|{header:^{col_key_to_width[key]+PADDING}}",end='')
    printed_width += col_key_to_width[key]+PADDING + 1 #extra 1 for the '|'
print("|")
printed_width+=1 #another |


print('-' * printed_width)

#the magic, dicts have fixed order, so we know it will always match,
#nested format strings allow us to dynamically size the fields based
#on the dict above
for student in students:
    for field,value in student.items():
        print(f"|{value:^{col_key_to_width[field]+PADDING}}",end='')
    print("|")


print('-' * printed_width)    


````

The multi-student entry loop, dictionary-driven dynamic column widths, and **all three bonus challenges** are present and working: Dean's List status, required GPA in remaining credits (with sensible N/A handling for out-of-range values), and the graduation-met case via `max(TOTAL_CREDITS - earned, 0)`.
