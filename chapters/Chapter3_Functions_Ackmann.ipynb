## 3.1 Why Functions?

You've already used functions like `print()`, `input()`, `len()`, and `type()`. Now you'll learn to create your own.

**Functions solve three problems:**

1. **Repetition:** Write code once, use it many times
2. **Organization:** Break complex programs into manageable pieces
3. **Abstraction:** Hide implementation details behind a simple interface

Consider this code that calculates the credit completion rate for three Crestview students:

````python
# Without functions - repetitive and error-prone
earned1 = 99
attempted1 = 105
rate1 = earned1 / attempted1 * 100
print(f"Student 1 completion rate: {rate1:.1f}%")

earned2 = 45
attempted2 = 52
rate2 = earned2 / attempted2 * 100
print(f"Student 2 completion rate: {rate2:.1f}%")

earned3 = 120
attempted3 = 125
rate3 = earned3 / attempted3 * 100
print(f"Student 3 completion rate: {rate3:.1f}%")
````

If the registrar changes how completion rate is calculated (say, excluding withdrawn courses), you'd have to update the formula in three places. And with 600 students in the full dataset? Six hundred places. With a function, you update it once.

## 3.2 Defining and Calling Functions

Use the `def` keyword to define a function:

```python
def function_name():
    # code to execute
    # indented block
```

The function doesn't run until you **call** it by using its name followed by parentheses.

````python
# Define a simple function
def greet():
    print("Welcome to the Crestview Advising Center!")
    print("Let's review your academic record.")

# Nothing happens until we call it
print("Before calling the function")
greet()  # Call the function
print("After calling the function")
````

````python
# You can call a function multiple times
def print_separator():
    print("=" * 40)

print_separator()
print("CRESTVIEW STUDENT GPA REPORT")
print_separator()
print("College of Business avg GPA: 3.27")
print("Engineering avg GPA: 3.31")
print_separator()
````

### Function Naming Conventions

Follow the same rules as variables:
- Use `snake_case` (lowercase with underscores)
- Start with a verb that describes what the function does
- Be descriptive but concise

**Good names:** `calculate_gpa()`, `get_student_record()`, `print_report()`, `validate_credits()`

**Bad names:** `func1()`, `do_stuff()`, `x()`, `CalculateGpa()` (that style is for classes)

## 3.3 Parameters and Arguments

Functions become powerful when they accept input. **Parameters** are variables defined in the function header. **Arguments** are the actual values you pass when calling the function.

````python
# Function with one parameter
def greet_student(name):  # 'name' is the parameter
    print(f"Hello, {name}! Your advising appointment is confirmed.")

greet_student("Maria")    # "Maria" is the argument
greet_student("James")
greet_student("Priya")
````

````python
# Function with multiple parameters
def show_quality_points(gpa, credits):
    quality_points = gpa * credits
    print(f"GPA {gpa} x {credits} credits = {quality_points:.1f} quality points")

show_quality_points(3.41, 99)
show_quality_points(2.85, 60)
show_quality_points(4.0, 15)
````

````python
# Arguments are matched by position
def describe_student(name, college, credits):
    print(f"{name} is in {college} with {credits} credits earned")

describe_student("Maria", "College of Business", 105)
describe_student("James", "Engineering", 45)
````

````python
# ⚠️ What happens if you mix up the order? Run it and look closely.
describe_student(105, "Maria", "College of Business")
````

**Read that output again: `105 is in Maria with College of Business credits earned`.**

Notice what did NOT happen: Python did not raise an error. The function ran "successfully" and produced nonsense. This is called a **silent failure**, and it's more dangerous than a crash. A crash stops you immediately; garbage output can flow into a report that reaches the Provost's office before anyone notices.

This is the most important lesson in this section: **Python matches arguments by position and trusts that you got the order right.** In section 3.8 you'll learn keyword arguments, which protect you from exactly this mistake.

## 3.4 Return Values

Most functions should **return** a value rather than print it. This lets you use the result in further calculations.

The `return` statement:
1. Sends a value back to the caller
2. Immediately exits the function

````python
# Function that returns a value
def calculate_completion_rate(earned, attempted):
    rate = earned / attempted
    return rate

# Capture the returned value
my_rate = calculate_completion_rate(99, 105)
print(f"Completion rate: {my_rate:.2%}")

# Use it directly in an expression
gap = 1 - calculate_completion_rate(99, 105)
print(f"Share of credits not completed: {gap:.2%}")
````

````python
# Compare: print vs return

# This function PRINTS but doesn't RETURN
def add_credits_and_print(fall, spring):
    total = fall + spring
    print(total)

# This function RETURNS but doesn't PRINT
def add_credits_and_return(fall, spring):
    total = fall + spring
    return total

# See the difference
x = add_credits_and_print(15, 12)   # Prints 27
print(f"x is: {x}")                  # x is None!

y = add_credits_and_return(15, 12)  # Nothing printed
print(f"y is: {y}")                  # y is 27
````

**Key insight:** If you don't explicitly return a value, Python returns `None` automatically.

**Best practice:** Functions should usually return values, not print them. Let the caller decide what to do with the result.

````python
# Return can exit early from a function
# This is the academic standing logic from Chapter 2, now reusable
def classify_standing(gpa):
    if gpa < 0 or gpa > 4.0:
        return "Invalid GPA"  # Exit immediately
    if gpa >= 3.75:
        return "Dean's List with Distinction"
    if gpa >= 3.5:
        return "Dean's List"
    if gpa >= 2.0:
        return "Good Standing"
    if gpa >= 1.5:
        return "Academic Probation"
    return "Academic Suspension"

print(classify_standing(3.65))
print(classify_standing(1.85))
print(classify_standing(3.90))
print(classify_standing(5.0))
````

> **A Quick Preview:** The next example returns **two** values at once and "unpacks" them into two variables. Python does this with a **tuple**, a data structure you'll study properly in **Chapter 4**. For now, just read the pattern: `return minimum, maximum` sends both values back, and `low, high = ...` catches them in order.

````python
# Functions can return multiple values using a tuple
def get_min_max(numbers):
    minimum = numbers[0]
    maximum = numbers[0]
    for num in numbers:
        if num < minimum:
            minimum = num
        if num > maximum:
            maximum = num
    return minimum, maximum  # Returns a tuple

gpas = [3.41, 3.55, 1.85, 3.90, 4.0]
low, high = get_min_max(gpas)  # Unpack the tuple
print(f"GPA range across students: {low} to {high}")
````

## 3.5 Variable Scope

**Scope** determines where a variable can be accessed. Python has two main scopes:

1. **Local scope:** Variables created inside a function exist only within that function
2. **Global scope:** Variables created outside all functions can be accessed anywhere

This is one of the most common sources of confusion for beginners.

````python
# Local variables exist only inside the function
def check_deans_list(gpa):
    threshold = 3.5  # Local variable
    return gpa >= threshold

result = check_deans_list(3.65)
print(f"Dean's List: {result}")

# This will cause an error - threshold doesn't exist here
# print(threshold)  # Uncomment to see the error
````

````python
# Global variables can be read inside functions
graduation_requirement = 120  # Global variable

def credits_remaining(earned):
    return graduation_requirement - earned  # Reading global variable - OK

print(credits_remaining(99))
````

````python
# Local variables shadow global variables with the same name
college = "College of Business"  # Global

def show_college():
    college = "Engineering"  # Local - different variable!
    print(f"Inside function: {college}")

show_college()
print(f"Outside function: {college}")  # Global unchanged
````

````python
# Common mistake: trying to modify a global without declaring it
students_processed = 0

def increment():
    # students_processed = students_processed + 1  # This would cause an error!
    # Python thinks you're creating a local variable
    # but referencing it before assignment
    pass

# If you really need to modify a global (usually you shouldn't):
def increment_global():
    global students_processed
    students_processed = students_processed + 1

print(f"Before: {students_processed}")
increment_global()
print(f"After: {students_processed}")
````

**Best practice:** Avoid using `global`. Instead, pass values as parameters and return results. This makes functions predictable and easier to test.

````python
# Better approach - pass and return values
def increment(value):
    return value + 1

students_processed = 0
students_processed = increment(students_processed)
students_processed = increment(students_processed)
students_processed = increment(students_processed)
print(f"Students processed: {students_processed}")
````

## 3.6 Putting It Together: Completion Rate Refactored

Let's revisit the repetitive code from section 3.1 with a proper function.

> **A Quick Preview:** The function below uses two features we haven't covered yet: a **default parameter value** (`standard=0.85`) and a **docstring** (the triple-quoted description). Both are coming up in the WALK section (3.7 and 3.9). For now, just notice how readable they make the function.

````python
# Clean, reusable solution
def meets_completion_standard(earned, attempted, standard=0.85):
    '''Return True if the student's completion rate meets the standard.'''
    rate = earned / attempted
    return rate >= standard

# Now it's easy to check many students
# (earned, attempted) for three students
records = [(99, 105), (45, 52), (30, 40)]

for earned, attempted in records:
    flag = meets_completion_standard(earned, attempted)
    print(f"{earned}/{attempted} credits - meets standard: {flag}")

# If the university raises the standard, change one argument:
print("\nWith a 90% standard:")
for earned, attempted in records:
    flag = meets_completion_standard(earned, attempted, 0.90)
    print(f"{earned}/{attempted} credits - meets standard: {flag}")
````

## 3.7 Default Parameter Values

You can give parameters default values. If the caller doesn't provide an argument, the default is used.

````python
def welcome_student(name, college="Undeclared"):
    return f"Welcome, {name}! You are registered under: {college}"

print(welcome_student("Maria"))                      # Uses default
print(welcome_student("James", "Engineering"))       # Overrides default
print(welcome_student("Priya", "College of Health"))
````

> **A Quick Preview:** The next function returns a **dictionary**, a data structure that stores labeled values (like a student record with named fields). Dictionaries are covered fully in **Chapter 4**. For now, read `{"name": name, ...}` as "a record with a name field, a gpa field," and so on.

````python
# Parameters with defaults must come after parameters without defaults
def create_student_record(name, gpa, college="Undeclared", full_time=True):
    return {
        "name": name,
        "gpa": gpa,
        "college": college,
        "full_time": full_time
    }

print(create_student_record("Maria", 3.41))
print(create_student_record("James", 3.55, "Engineering"))
print(create_student_record("Priya", 3.90, "College of Health", False))
````

## 3.8 Keyword Arguments

You can specify arguments by name, which makes function calls clearer and lets you skip default values.

Remember the silent failure from section 3.3, where `describe_student(105, "Maria", ...)` produced garbage? Keyword arguments are the protection against that mistake: when you name each argument, order no longer matters.

````python
def format_gpa(gpa, decimals=2, show_scale=True):
    if show_scale:
        return f"{gpa:.{decimals}f} / 4.0"
    else:
        return f"{gpa:.{decimals}f}"

# Positional arguments (order matters)
print(format_gpa(3.41666))

# Keyword arguments (order doesn't matter)
print(format_gpa(3.41666, decimals=1))
print(format_gpa(3.41666, show_scale=False, decimals=3))
print(format_gpa(decimals=2, gpa=3.41666))  # Even the first one can be named
````

````python
# Keyword arguments make complex function calls readable
def send_advising_email(to, subject, body, cc=None, urgent=False):
    print(f"To: {to}")
    print(f"Subject: {'[URGENT] ' if urgent else ''}{subject}")
    if cc:
        print(f"CC: {cc}")
    print(f"Body: {body}")

# Clear what each argument means
send_advising_email(
    to="maria@crestview.edu",
    subject="Midterm Grade Alert",
    body="Please schedule an advising appointment this week.",
    cc="advisor@crestview.edu",
    urgent=True
)
````

## 3.9 Docstrings

A **docstring** is a string that documents what a function does. It appears right after the `def` line and uses triple quotes.

Good docstrings describe:
1. What the function does
2. What parameters it expects
3. What it returns

````python
def calculate_required_gpa(current_gpa, credits_completed, target_gpa, total_credits):
    """
    Calculate the GPA needed in remaining credits to reach a target overall GPA.

    Parameters:
        current_gpa: Student's current cumulative GPA
        credits_completed: Credits completed so far
        target_gpa: Desired overall GPA at total_credits
        total_credits: Total credits at which the target should be reached

    Returns:
        The GPA required in the remaining credits.
        Note: a result above 4.0 means the target is not achievable.

    Example:
        >>> calculate_required_gpa(3.0, 90, 3.2, 120)
        3.8
    """
    remaining = total_credits - credits_completed
    required = (target_gpa * total_credits - current_gpa * credits_completed) / remaining
    return round(required, 2)

# Now you can see the docstring with help()
help(calculate_required_gpa)
````

````python
# Use the function: a junior with a 3.0 wants a 3.2 by graduation
required = calculate_required_gpa(3.0, 90, 3.2, 120)
print(f"Required GPA in remaining credits: {required}")

# Is a 3.8 over their final 30 credits realistic? That's an advising conversation.
````

## 3.10 Input Validation

Professional functions validate their inputs and handle edge cases gracefully.

**This is your Week 3 toolkit.** The messy Crestview dataset contains missing GPAs, text where numbers should be, and impossible values. Validation functions like the one below are how you'll survive it.

````python
def calculate_average_gpa_safe(gpas):
    """
    Calculate the average of a list of GPAs.
    Returns None if the list is empty or invalid.
    """
    # Validate input
    if not gpas:  # Empty list or None
        return None

    if not isinstance(gpas, list):
        return None

    # Calculate
    total = 0
    count = 0
    for gpa in gpas:
        if isinstance(gpa, (int, float)):  # Only process numbers
            total += gpa
            count += 1

    if count == 0:
        return None

    return total / count

# Test with various inputs
print(calculate_average_gpa_safe([3.41, 3.55, 2.10]))   # Normal case
print(calculate_average_gpa_safe([]))                    # Empty list
print(calculate_average_gpa_safe(None))                  # None
print(calculate_average_gpa_safe([3.5, "N/A", 2.5]))     # Messy data!
````

````python
# Raising exceptions for invalid input
def calculate_completion_rate(earned, attempted):
    """
    Calculate credit completion rate.
    Raises ValueError if attempted is zero.
    """
    if attempted == 0:
        raise ValueError("Cannot calculate a completion rate for 0 attempted credits")
    return earned / attempted

print(calculate_completion_rate(99, 105))

# ⚠️ Uncomment to see the exception
#print(calculate_completion_rate(0, 0))
````

## 3.11 Functions Calling Functions

Complex programs are built by combining simple functions. Each function should do one thing well.

````python
# Small, focused functions
def calculate_quality_points(gpa, credits):
    """Calculate quality points (GPA x credits)."""
    return gpa * credits

def calculate_completion_rate(earned, attempted):
    """Calculate credit completion rate."""
    return earned / attempted

def classify_standing(gpa):
    """Classify academic standing from GPA."""
    if gpa >= 3.5:
        return "Dean's List"
    elif gpa >= 2.0:
        return "Good Standing"
    else:
        return "Academic Probation"

# Combine them into a higher-level function
def summarize_student(student_id, gpa, earned, attempted):
    """
    Build a one-line academic summary for a student.
    """
    points = calculate_quality_points(gpa, earned)
    rate = calculate_completion_rate(earned, attempted)
    standing = classify_standing(gpa)
    return f"{student_id}: GPA {gpa:.2f} | {points:.1f} quality pts | {rate:.1%} completion | {standing}"

# Use the high-level function
print(summarize_student("CU100001", 3.41, 99, 105))
print(summarize_student("CU100003", 1.85, 38, 45))
````

````python
# Your code here

#I'm using type annotations just to keep track of my understanding of the problem
def calculate_gpa(letter_grades: list[str] | str) -> float:
    """
    Calculate the GPA based on a list of letter grades.

    Parameters:
        letter_grades: Letter grades for a student
                       For ease of use this can be a single letter grade as a string 'B+'
                       or a list of letter grades ['A', 'B+']
                       
        
    Returns:
        The GPA resulting from the input list.
        

    Example:
        >>> calculate_gpa(['A','B'])
        3.5

    Notes:
        Invalid grades are getting dropped at this point since we don't know the incoming dataset enough to know
        all the goofy ways data might be entered.
    """
    if type(letter_grades) == str:
        letter_grades = [letter_grades]
    
    letter_to_gpa = { 'A': 4.0, 'A-':3.7, 'B+':3.3, 'B':3.0, 'B-':2.7, 'C+':2.3, 'C':2.0, 'C-':1.7, 'D':1.0, 'F':0.0 }
    gpas = []
    for grade in letter_grades:
        grade = grade.upper()
        if grade in letter_to_gpa:
            gpas.append(letter_to_gpa[grade])

    #protect the all invalid or empty list
    if len(gpas) == 0:
        return 0.0
    return sum(gpas)/len(gpas)
    

def check_gpa(actual, expected):
    assert calculate_gpa(actual) == expected, f"Expected {expected!r}, got {actual!r}"
    print(f"✓ {expected!r}")

check_gpa('B+', 3.3)
check_gpa(['B+'], 3.3)
check_gpa(["a","b",'x'],3.5)
check_gpa([],0.0)
check_gpa(['x','x','A++'],0.0)
check_gpa(['A','B','C'],3.0)
````

### Problem 3.8: GPA Data Cleaner (Week 3 Warm-Up)
The registrar sends you a GPA list with data entry errors. Write a set of functions to process it:

1. `validate_gpa(value)` - Returns True if value is a number between 0.0 and 4.0 (inclusive)
2. `classify_standing(gpa)` - Returns "Dean's List" (>= 3.5), "Good Standing" (>= 2.0), or "Academic Probation" (below 2.0)
3. `process_gpas(gpa_list)` - Takes a list of raw values, validates each one, and returns a dictionary with:
   - "valid_gpas": list of valid GPAs
   - "average_gpa": average of the valid GPAs
   - "deans_list_count": how many valid GPAs qualify for Dean's List

Test with: `[3.5, -1.0, 2.8, "N/A", 4.0, 5.0, 1.9]`
(Expected: 4 valid GPAs, average 3.05, 2 on Dean's List)

````python
# Your code here
raw_gpas = [3.5, -1.0, 2.8, "N/A", 4.0, 5.0, 1.9]

def validate_gpa(value:float) -> bool:
    if type(value) != float:
        return False
    return value >= 0.0 and value <= 4.0

def classify_standing(gpa: float) -> str:
    if gpa >= 3.5: 
        return "Dean's List"
    elif gpa >= 2.0:
        return "Good Standing"
    else:
        return "Academic Probation"

def process_gpas(gpa_list) -> dict:
    res = {"valid_gpas":[], "average_gpa":0.0,  "deans_list_count":0}
    for gpa in gpa_list:
        if validate_gpa(gpa):
            res["valid_gpas"].append(gpa)
            res["average_gpa"] += gpa
            if classify_standing(gpa) == "Dean's List":
                res["deans_list_count"]+= 1

    if len(res["valid_gpas"]) > 1:
        res["average_gpa"] = res["average_gpa"] / len(res["valid_gpas"])
    return res

print(process_gpas(raw_gpas))
````

````python
# STUDENT GRADE ANALYZER PROJECT
#
# Document your AI collaboration below:
# - Prompts used:
# - Key insights from AI:
# - Modifications you made:
#
# Your code below:

# Sample data
students = [
    {"name": "Alice", "college": "Business", "gpa": 3.8, "credits": 45},
    {"name": "Bob", "college": "Engineering", "gpa": 3.2, "credits": 52},
    {"name": "Charlie", "college": "Arts", "gpa": 2.9, "credits": 38},
    {"name": "Diana", "college": "Business", "gpa": 3.95, "credits": 60},
    {"name": "Eve", "college": "Engineering", "gpa": 3.5, "credits": 48},
    {"name": "Frank", "college": "Health", "gpa": 3.1, "credits": 55},
    {"name": "Grace", "college": "Arts", "gpa": 3.7, "credits": 42},
    {"name": "Henry", "college": "Health", "gpa": 2.5, "credits": 30},
]

# Start building your functions here...

def get_letter_grade(gpa):
    """Convert GPA to letter grade."""
    #Convert GPA to letter grade (A: 3.7+, B: 3.0-3.69, C: 2.0-2.99, D: 1.0-1.99, F: below 1.0)
    if gpa >= 3.7:
        return 'A'
    elif gpa >= 3.0:
        return 'B'
    elif gpa >= 2.0:
        return 'C'
    elif gpa >= 1.0:
        return 'D'
    else:
        return 'F'


def calculate_class_standing(credits):
    """Convert Class Year Standing."""
    # - Return "First Year" (<30), "Sophomore" (30-59), "Junior" (60-89), "Senior" (90+)
    if credits >= 90:
        return 'Senior'
    elif credits >= 60:
        return 'Junior'
    elif credits >= 30:
        return 'Sophomore'
    else:
        return 'First Year'
    

def is_deans_list(gpa, credits):
    """Returns whether the gpa/credit combo is deans list worthy."""
    # - Return True if GPA >= 3.5 AND credits >= 12
    return gpa >= 3.5 and credits >= 12
    

def filter_by_college(students, college):
    """Returns only return studens in a specific college."""
    # - Return list of students in a specific college
    return [s for s in students if s['college'].casefold() == college.casefold() ]

def calculate_college_stats(students, college):
    
    """Caclulate college stats."""
    ret = {'college':college.capitalize(), 'count': 0, 'avg_gpa':0, 'deans_list_count':0}

    for student in filter_by_college(students,college):
        ret['count'] += 1
        ret['avg_gpa'] += student['gpa']
        if is_deans_list(student['gpa'], student['credits']):
            ret['deans_list_count'] += 1

    if ret['count'] > 1:
        ret['avg_gpa'] = ret['avg_gpa'] /ret['count'] 
    return ret

#i added a default parameter for print width, sorry...
def generate_report(students, width = 80):
    """Generates the final output report"""
    # - Print a formatted report showing:
    colleges = set()
    # Each student's name, college, GPA, letter grade, and standing
    # Summary statistics by college
    # List of dean's list students
    print("=" * width)
    print(f"{'CRESTVIEW UNIVERSITY - STUDENT GRADE ANALYZER':^{width}}")
    print("=" * width)

    print()
    print("STUDENT INFORMATION")
    print("-" *width)

    #2 spaces so equal widths are (80 - ((4cols -1) * 2))/4 = 18.5
    col_width = [ 18, 26, 12, 18]

    print(f"{'Student Name':<{col_width[0]}}  {'College':<{col_width[1]}}  {'Grade':^{col_width[2]}}  {'Standing':>{col_width[3]}}")
    print(f"{'-' * col_width[0]}  {'-' * col_width[1]}  {'-' * col_width[2]}  {'-' * col_width[3]}")

    for student in students:
        colleges.add(student['college'].casefold())
        formatted_gpa = f"{student['gpa']:.2f} ({get_letter_grade(student['gpa'])})"
        print(f"{student['name']:<{col_width[0]}}  {student['college']:<{col_width[1]}}  {formatted_gpa:^{col_width[2]}}  {calculate_class_standing(student['credits']):>{col_width[3]}}")
    print()
    print(f"{len(students)} Total Students...")
   
    print()
    print()
    print(f"COLLEGE SUMMARY")
    print("-" *width)  
    col_width = [ 24, 18, 12, 20]
    print(f"{'College':<{col_width[0]}}  {'Student Count':^{col_width[1]}}  {'Avg. GPA':^{col_width[2]}}  {"Dean's List Count":>{col_width[3]}}")
    print(f"{'-' * col_width[0]}  {'-' * col_width[1]}  {'-' * col_width[2]}  {'-' * col_width[3]}")
    for college in colleges:
        c_info = calculate_college_stats(students, college)
        print(f"{c_info['college']:<{col_width[0]}}  {c_info['count']:^{col_width[1]}}  {c_info['avg_gpa']:^{col_width[2]},.2f}  {c_info['deans_list_count']:>{col_width[3]}}")  

    print()
    print()
    print(f"DEAN'S LIST STUDENTS")
    print("-" *width)  
    for student in students:
        if is_deans_list(student['gpa'],student['credits']):
            print(student['name'])
    
 
    print()
    print("=" * width)
    print(f"{'END OF REPORT':^{width}}")
    print("=" * width)



generate_report(students,80)

````
